# RAG Fundamentals

Say I want to take a product description and then map it against a set of predefined categories, product categories. And assume that the list of product categories is hundreds of items long from say a standardized nomenclature. What RAG is going to help with is to do a pre-filter of the categories that we would then ask an LLM to apply the most relevant ones to the product description. The reason why this is helpful is we can't supply hundreds of categories to the large language model and ask it to map it against the product description as A) that's more expensive, but B) there's a risk of it missing the product category that's applied because it's inundated with a lot of noise. So RAG in this case helps us refine the list as a pre-filter before we get the large language model to define what the category should be.
