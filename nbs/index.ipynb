# casblog

> 

Early stages implementation of my blog using FastHTML within Solveit as an nbdev project

Because this is deployed with [pla.sh](https://pla.sh/), we had to have the primary file named `main.py`. I could have just made the primary notebook `main.py` but that would mean running the `plash_deploy --force_data` command not in the project root, which complicated where we put the `/data` directory (used in dev and prod).

Note: right now we are forcing `/data` dir update with each deploy via `plash_deploy --force_data`. This is a temp solution, as this means we are keeping the source of truth prod db within solveit.
